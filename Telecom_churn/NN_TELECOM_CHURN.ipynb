```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import tensorflow as tf
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
```

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
```

```python
df=pd.read_csv('telecom_churn.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df_num=df.select_dtypes(['float64','int64'])
```

```python
df_num.head()
```

```python
df_cat=df.select_dtypes(object)
```

```python
df_cat.head()
```

```python
df_cat["TotalCharges"].replace(' ',np.nan,inplace=True)
```

```python
df_cat['TotalCharges'].isnull().sum()
```

```python
df_cat["TotalCharges"]=df_cat["TotalCharges"].astype("float")
```

```python
df_cat["TotalCharges"].fillna(df_cat["TotalCharges"].mean(),inplace=True)
```

```python
df_num["TotalCharges"]=df_cat["TotalCharges"]
```

```python
df_num.head()
```

```python
df_cat.drop('TotalCharges',axis=1,inplace=True)
```

```python
df_cat.head()
```

```python
for col in df_cat:
    le=LabelEncoder()
    df_cat[col]=le.fit_transform(df_cat[col])
```

```python
df_cat.head()
```

```python
df_new=pd.concat([df_cat,df_num],axis=1)
```

```python
df_new.head()
```

```python
df_new.shape
```

```python
df_new.drop('customerID',inplace=True,axis=1)
```

```python
X=df_new.drop('Churn',axis=1)
Y=df_new["Churn"]
```

```python
X.head()
```

```python
Y.head()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
ss=StandardScaler()
```

```python
x_train_ss=ss.fit_transform(X_train)
x_test_ss=ss.transform(X_test)
```

```python
X.shape
```

```python
model=tf.keras.Sequential([
       tf.keras.layers.Dense(2,activation="relu",input_shape=(X.shape[1],)),
       tf.keras.layers.Dense(3,activation="relu"),
       tf.keras.layers.Dense(1,activation="sigmoid")
])
```

```python
model.summary()
```

```python
model.compile(optimizer='sgd',loss='binary_crossentropy')
```

```python
trained_model=model.fit(x_train_ss,Y_train,epochs=50,batch_size=32)
```

```python
plt.plot(trained_model.history['loss'])
```

```python
Y_pred=model.predict(x_test_ss)
```

```python
Y_pred
```

```python
Y_pred=np.where(Y_pred >= 0.5,1,0)
```

```python
print(classification_report(Y_test,Y_pred))
```

```python
#ROS
```

```python
model1=tf.keras.Sequential([
       tf.keras.layers.Dense(2,activation="relu",input_shape=(X.shape[1],)),
       tf.keras.layers.Dense(3,activation="relu"),
       tf.keras.layers.Dense(1,activation="sigmoid")
])
```

```python
model1.summary()
```

```python
model1.compile(optimizer='sgd',loss='binary_crossentropy')
```

```python
ros=RandomOverSampler()
```

```python
X_ros,Y_ros=ros.fit_sample(x_train_ss,Y_train)
```

```python
trained_model1=model1.fit(X_ros,Y_ros,epochs=50,batch_size=32)
```

```python
plt.plot(trained_model1.history['loss'])
plt.show()
```

```python
Y_pred=model1.predict(x_test_ss)
```

```python
Y_pred=np.where(Y_pred >= 0.5,1,0)
```

```python
print(classification_report(Y_test,Y_pred))
```
